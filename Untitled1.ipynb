```python
import pdfplumber
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with pdfplumber.open(r'Kaustubh Pagar Resume.pdf') as pdf:
    first_page = pdf.pages[0]
    resume_text = first_page.extract_text()

stop_words = set(stopwords.words('english'))
  
word_tokens = word_tokenize(resume_text)

filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
  
filtered_sentence = []
  
for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
  
```

```python
def listToString(s):
   
    
    s1 = " "
   
     
    return (s1.join(s))
       
         
resume_text_wo_stopwords = listToString(filtered_sentence)
```

```python
wnl = WordNetLemmatizer()

lemmatized_resume_text = wnl.lemmatize(resume_text_wo_stopwords)

tfidf = TfidfVectorizer()
```

```python
job_skills = "Data Science,Data Analysis,Database,SQL,Machine Learning,tableau"
```

```python
import re 
```

```python
# Extract skills from a document
def extract_skills_from_document():
  skills_from_doc_endpoint = "https://emsiservices.com/skills/versions/latest/extract"
  text = lemmatized_resume_text
  confidence_interval = '1'
  payload = "{ \"text\": \"... " + text + " ...\", \"confidenceThreshold\": " + confidence_interval + " }"

  headers = {
      'authorization': "Bearer " + access_token,
      'content-type': "application/json"
      }

  response = requests.request("POST", skills_from_doc_endpoint, data=payload.encode('utf-8'), headers=headers)
    
  

  skills_found_in_document_df = pd.DataFrame(json_normalize(response.json()['data']));
  skills_from_res = skills_found_in_document_df['skill.name']# Where response is a JSON object drilled down to the level of 'data' key
  skill_text = skills_from_res 
  skill_str = skills_from_res.to_string(index=False)
  final_skill_str = re.sub("[\(\[].*?[\)\]]", "", skill_str)
  print(final_skill_str)
  corpus = [final_skill_str, job_skills]
  
  tfidf_matrix = tfidf.fit_transform(corpus)
    
  cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
  print("Match =",cosine_sim[0][1]*100,"%")
  

  
  
  


  
extract_skills_from_document()
```
